## Defining a user-defined function with if-else conditions

```python
# function definition
def price_label(row):
    if row['prices'] <= 5:
        return 'low_range_product'
    elif (row['prices'] > 5 and row['prices'] <= 15):
        return 'mid_range_product'
    elif row['prices'] > 15:
        return 'high_range_product'
    else: return 'Not enough data'
```

```python
# calling the function b creating a new column in the df dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

### If-Statements with For-Loops

##### To find out how busy each day of the week is?

```python
df_ords_prods_merge['order_days_of_week'].value_counts()
```

```python
# create a list to store the results from the for-loop
result = []
```

## Classifying the hours of the day into 3 categories
#### Most_orders     :- contains top 8 hours having highest value counts (10,11,14,15,13,12,16,9)
#### Average_orders  :- contains middle 8 hours (17,8,18,19,20,7,21,22)
#### Fewest_orders   :- contais the last 8 hours with lowest value counts (23,6,0,1,5,2,4,3)

```python
# use If-Statemenst with for-loop

result2 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result2.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result2.append("Fewest orders")
  else:
    result2.append("Average orders")
```

```python
#printing the result
result2
```

```python
#create a new column named busiest_period_of_day and assign the value of result2 to it
df_ords_prods_merge['busiest_period_of_day'] = result2
```

```python
#Print the frequency for this new column
df_ords_prods_merge['busiest_period_of_day'].value_counts() 
```
