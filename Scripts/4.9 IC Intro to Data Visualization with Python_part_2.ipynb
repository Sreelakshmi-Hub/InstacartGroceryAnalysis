# Steps 1 and 2 : Importing Libraries

```python
# import the required libraries for analysis
import pandas as pd
import numpy as np
import os
```

```python
# create a path variable
path = r'C:\Users\Sreelakshmi\Desktop\Instacart Basket Analysis'
```

## Step 3: Import the new dataset customer to a dataframe

```python
# importing the new dataset
df_cust = pd.read_csv(os.path.join(path,'Data', 'Original Data', 'customers.csv'), index_col = False)
```

```python
# displaying the contents of the dataframe
df_cust.head()
```

## Step 4: Data Wrangling on the dataframe df_cust

### 1. Dropping unwanted columns 

```python
# Dropping the columns date_joined as it is not required for the analysis
df_cust = df_cust.drop(columns = ['date_joined'])
```

### 2. Renaming column names for consistency

```python
# renaming columns
df_cust.rename(columns = {'First Name' : 'first_name'}, inplace = True)
df_cust.rename(columns = {'Surnam' : 'surname'}, inplace = True)
df_cust.rename(columns = {'Gender' : 'gender'}, inplace = True)
df_cust.rename(columns = {'STATE' : 'state'}, inplace = True)
df_cust.rename(columns = {'Age' : 'age'}, inplace = True)
df_cust.rename(columns = {'fam_status' : 'family_status'}, inplace = True)
```

```python
#displaying the contents of the dataframe
df_cust.head()
```

```python
# checking for exploratory data analysis
df_cust.describe()
```

## Step 5 : Data consistency checks on the dataframe df_cust

### 1. Checking for missing values in the dataframe

```python
#check for missing values
df_cust.isnull().sum()
```

```python
# Extracting records that has first_name with null values
df_nan = df_cust[df_cust['first_name'].isnull() == True]
```

```python
# output the contents
df_nan
```

```python
# Total rows in the dataframe
df_cust.shape
```

#### Here only 5% of the dataframe contains missing values in the column first_name. It is better to leave the nulls as it is than deleting all the records with first_name with null values, as they don't affect the analysis.

### 2. Checking for mixed datatypes in the dataframe

```python
# check for mixed types
for col in df_cust.columns.tolist():
    wierd = (df_cust[[col]].map(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
    if len(df_cust[wierd]) > 0:
        print(col)
```

```python
df_cust['first_name'].value_counts(dropna = False)
```

### Addressing mixed datatype columns by changing its datatype

```python
# setting the datatype of column first_name to string
df_cust['first_name'] = df_cust['first_name'].astype('str')
```

### 3. Checking for duplicates in the dataframe

```python
# checking duplicates in the dataframe
df_dups = df_cust[df_cust.duplicated()]
```

```python
df_dups
```

### No duplicate records have been found!!

## Step 6 : Combine the customer data with the rest of your prepared Instacart data

```python
# Import the prepared dataset ords_prods_grouped_agg.pkl to the dataframe
ords_prods_grouped_agg = pd.read_pickle(os.path.join(path,'Data', 'Prepared Data', 'ords_prods_grouped_agg.pkl'))
```

```python
ords_prods_grouped_agg.head()
```

```python
#checking the datatypes of all columns in ords_prods_grouped_agg
ords_prods_grouped_agg.info()
ords_prods_grouped_agg.shape
```

```python
#checking the datatypes of all columns in df_cust
df_cust.info()
df_cust.shape
```

```python
#merge the datasets ords_prods_grouped_agg and df_cust
df_merge = ords_prods_grouped_agg.merge(df_cust, on = 'user_id', indicator = 'Check')
```
